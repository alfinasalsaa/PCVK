# **Transformasi Linier Brightness**

```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

print(' Mengubah tingkat kecerahan Citra')
print('----------------------------------')
try:
  brightness = int(input('Masukkan nilai kecerahan: '))
except ValueError:
  print('Error, not a number')

original = cv.imread('/content/drive/MyDrive/PCVK/Images/female.tiff')
brightness_image = np.zeros(original.shape, original.dtype)

#akses per pixel
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      brightness_image[y,x,c] = np.clip(original[y,x,c] + brightness, 0, 255)

#cara simple tanpa for loop

final_frame = cv.hconcat((original, brightness_image))
cv2_imshow(final_frame)
```

#  **4. Implementasi Transformasi Grayscale**

```python
# A. Averaging
import cv2
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

print(' Transformasi grayscale dengan metode Averaging')
print('-----------------------------------------------')

image_path = ('/content/drive/MyDrive/PCVK/Images/KTP.png')
image = cv2.imread(image_path)

# Menghitung citra grayscale menggunakan metode averaging
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Menampilkan citra asli dan citra grayscale
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Citra Asli')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('on')

plt.subplot(1, 2, 2)
plt.title('Citra Grayscale Averaging')
plt.imshow(gray_image, cmap='gray')
plt.axis('on')

plt.show()
```

```python
# Luminance
import cv2
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

print('Transformasi Grayscale - Metode Luminance')
print('-----------------------------------------')

# Baca citra RGB
image = cv2.imread('/content/drive/MyDrive/PCVK/Images/KTP.png')

# Ambil komponen R, G, dan B dari citra
R, G, B = cv2.split(image)

# Hitung nilai maksimum dan minimum dari setiap komponen
max_value = np.maximum(np.maximum(R, G), B)
min_value = np.minimum(np.minimum(R, G), B)

# Hitung citra grayscale menggunakan rumus Max-Min Lightness
gray_image = (max_value + min_value) / 2

plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.title('Citra Asli')
plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
plt.axis('on')  # Menghilangkan skala

# Tampilkan citra grayscale
plt.subplot(122)
plt.title('Citra Grayscale Luminance')
plt.imshow(grayscale_image, cmap='gray')
plt.axis('on')  # Menghilangkan skala
plt.show()


```

# **5. Tampilkan warna tertentu**

```python
# Luminance
import cv2
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

print('Menampilkan warna tertentu pada Citra')
print('-------------------------------------')

original = cv2.imread('/content/drive/MyDrive/PCVK/Images/KTP.png')

lower_blue = np.array([0, 0, 100], dtype=np.uint8)
upper_blue = np.array([100, 100, 255], dtype=np.uint8)

# Buat maska untuk bagian biru pada citra
blue_mask = cv2.inRange(image, lower_blue, upper_blue)

# Buat maska invers
inverse_blue_mask = cv2.bitwise_not(blue_mask)

# Ubah bagian biru menjadi grayscale
blue_part = cv2.bitwise_and(image, image, mask=blue_mask)
blue_part_gray = cv2.cvtColor(blue_part, cv2.COLOR_BGR2GRAY)

# Bagian lain menjadi grayscale
inverse_part = cv2.bitwise_and(image, image, mask=inverse_blue_mask)
inverse_part_gray = cv2.cvtColor(inverse_part, cv2.COLOR_BGR2GRAY)

# Gabungkan hasilnya
result = cv2.addWeighted(blue_part_gray, 1, inverse_part_gray, 1, 0)

# Tampilkan citra hasil
cv2_imshow(result)
cv2.waitKey(0)
cv2.destroyAllWindows()


```
